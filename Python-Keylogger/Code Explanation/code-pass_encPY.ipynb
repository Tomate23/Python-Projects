## Full code (pass_enc.py)
Basically what this code does is encrypt our password and generate a private key which we will use to decrypt our password, to do this we use the ``cryptography.fernet`` module from python.

```python
from cryptography.fernet import Fernet

def generar_key():
    key = Fernet.generate_key()
    with open('pass.key', 'wb') as file:
        file.write(key)

def cargar_key():
    return open('pass.key', 'rb').read()

generar_key()

key = cargar_key()

password = b'roma00josue23'

file = open('pass.enc', 'wb')

clave = Fernet(key)
pass_enc = clave.encrypt(password)
file.write(pass_enc)
file.close()
```

````python
# Fernet module is imported from the 
# cryptography package 
from cryptography.fernet import Fernet 

# key is generated 
key = Fernet.generate_key() 

# value of key is assigned to a variable 
f = Fernet(key) 
print(f"Our private key generated is > {f}\n")

# the plaintext is converted to ciphertext 
token = f.encrypt(b"welcome to geeksforgeeks") # IMPORTANT: the plain text in >>bytes<<

# display the ciphertext 
print(f"Encrypted token > {token}\n")

# decrypting the ciphertext 
d = f.decrypt(token) 

# display the plaintext 
print(d) 
````
